Buy 1 Get 1 Case Study

 

Mr. Bruce is the owner of Wayne group of superstores. He looks at the 3 months historical sales data of his stores and finds the sales to be dull. He approaches Queen Consultancy firm to solve this issue. You are working as a Data Scientist in that firm.

 

Mr. Bruce tells you that he wants a plan of action where he can give an offer like “Buy 1 Get 1 Free”. You need to perform Market Basket optimization using Apriori algorithm and give him the solution.

```python
from apyori import apriori
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
grocery= pd.read_csv('groceries.csv', header= None)
grocery
```

```python
grocery.shape
```

```python
# We will convert every transaction into a list ---> for loop
transactions = []
for i in range(0,len(grocery)):
    transactions.append([str(grocery.values[i,j]) for j in range(0, len(grocery.columns))])
```

```python
transactions
```

```python
from apyori import apriori

# we will create the a set of rules with this function
rules = apriori(transactions, min_support = 0.00213, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 3)
# transactions is my data

# Min_support, min_confidence and min_lift is up to you to decide

# min_length and max_length -> it will depend upon the type of offers you are planning to launch.
```

```python
# we will store the rules as a list and store in results 
results = list(rules)
print(*results, sep='\n')
```

```python
def conversion(results):
    lhs = [tuple(i[2][0][0])[0] for i in results]
    rhs = [tuple(i[2][0][1])[0] for i in results]
    support = [i[1] for i in results]
    confidence = [i[2][0][2] for i in results]
    lift = [i[2][0][3] for i in results]
    return list(zip(lhs,rhs,support,confidence,lift))
results_df = pd.DataFrame(conversion(results),columns=['LHS','RHS','Support','Confidence','Lift'])
```

```python
results_df.sort_values(by= 'Lift', ascending= False )
```
